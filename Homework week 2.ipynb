## The Orbit

progamming the orbit of the funtion.

$$F(x)=\left \{ \begin{array}{11}
   2x& 0\leq x < \frac{1}{2}\\
   \\ 2x-1 & \frac{1}{2} \leq x <1
   \end{array}
\right.$$


```python
#librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
    if(0<=x and x<0.5):
        return 2*x
    if(0.5<=x and x<1):
        return 2*x-1
```

```python
n=100

orbita1=np.array([0]*n, dtype= float)
orbita2=np.array([0]*n, dtype= float)
orbita3=np.array([0]*n, dtype= float)
orbita4=np.array([0]*n, dtype= float)
orbita5=np.array([0]*n, dtype= float)
orbita6=np.array([0]*n, dtype= float)
orbita7=np.array([0]*n, dtype= float)

orbita1[0]=0.1
orbita2[0]=0.142593
orbita3[0]=0.49999999
orbita4[0]=0.59999999
orbita5[0]=0.99999
#np.e/np.pi
```

```python
for i in range(n-1):
    orbita1[i+1]=F(orbita1[i])
    orbita2[i+1]=F(orbita2[i])
    orbita3[i+1]=F(orbita3[i])
    orbita4[i+1]=F(orbita4[i])
    orbita5[i+1]=F(orbita5[i])
```

### a) Plot the results of two cases in a plane $x_{i}$ vs $x_{i+1}$

```python
x=np.delete(orbita1,-1)
y=np.delete(orbita1,0)

#fig,limit1 = plt.subplots(1,1)
plt.plot(x,y,color='pink',label="semilla 1", marker='^')

x=np.delete(orbita2,-1)
y=np.delete(orbita2,0)
#fig,limit2=plt.subplots(1,1)
plt.plot(x,y,color='blue',label="semilla 2", marker='s')

x=np.delete(orbita3,-1)
y=np.delete(orbita3,0)
#fig,limit3=plt.subplots(1,1)
plt.plot(x,y,color='green',label="semilla 3", marker='d')

x=np.delete(orbita4,-1)
y=np.delete(orbita4,0)
#fig,limit4=plt.subplots(1,1)
plt.plot(x,y,color='black',label="semilla 4", marker='p')

x=np.delete(orbita5,-1)
y=np.delete(orbita5,0)
#fig,limit5=plt.subplots(1,1)
plt.plot(x,y,color='gray',label="semilla 5", marker='h')


for i in range (n-1):
    plt.plot(orbita1[i],orbita1[i+1],'^')
    plt.plot(orbita2[i],orbita2[i+1],'s')
    plt.plot(orbita3[i],orbita3[i+1],'d')
    plt.plot(orbita4[i],orbita4[i+1],'p')
    plt.plot(orbita5[i],orbita5[i+1],'h')
    
plt.legend()    
plt.title("Results of $x_i$ vs $x_{i+1}$")
plt.xlabel("Values $x_{i}$")
plt.ylabel("Values $x_{i+1}$")
plt.plot()
```

### b) Do they have a visible pattern? Do all (or almost all) orbits behave in the same way?

One can almost distinguish a pattern that is not very visible, depending on the value given to the seeds, a pattern that is almost similar is generated.

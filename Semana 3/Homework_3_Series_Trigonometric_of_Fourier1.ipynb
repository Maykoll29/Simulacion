The Fourier Trigonometric Series of an integrable function $f (x)$ on $(0, P)$ is written as

$$f(x) \thicksim \frac{a_0}{2}+\sum\limits_{k=1}^{n}[ a_{k}cos(\frac{2\pi k x}{P})+ b_k sin(\frac{2\pi k x}{P})]$$
Where
$$a_{0}=\frac{2}{P}\int_{0}^{P}f(x) dx $$
$$a_{k}=\frac{2}{P}\int_{0}^{P}f(x) cos(\frac{2\pi k x}{P}) dx; k\ge 1$$
$$b_{k}=\frac{2}{P}\int_{0}^{P}f(x) sin(\frac{2\pi k x}{P}) dx$$
Write a Python function $FourierSeries(f,P,x,n)$ that returns the values of the Fourier series of the function $f(x)$ at the point $x$. Plot the Fourier series of the following functions
$$f(x)=1-x$$
$$f(x)=x^{2},$$
using $P=1$, but calculate the series on the interval $(-2P,2P)$. Repeat the plot for $n$=6, 12, 36, 124. What do you observe as $n$ become larger?

### for: $1-x$

```python
#libraries
import numpy as np
from sympy import *
import matplotlib.pyplot as plt

from sympy import integrate, init_printing
from sympy.abc import x
init_printing(use_latex="mathjax")
from sympy import cos,pi,sin,tan
import sympy as sp
```

```python
P=1
def f(x):
    return 1-x
```

```python
#range for x 
x = np.linspace(-0.1111, 1+0.1, 1000)
```

```python
#define $a_0$
def a(n,P,y = 10000):
    a, b = 0, P
    dx = (b-a)/y
    varia = 0
    for x in np.linspace(a,b,y):
        varia += f(x) 
    varia *= dx
    return (2 / P) * varia
```

```python
#define $a_k$
def a_k(n,P,y=10000):
    a, b = 0, P
    dx = (b-a)/y
    varia = 0
    for x in np.linspace(a,b,y):
        varia += f(x)*np.cos((n*np.pi*x)/P)
    varia *= dx
    return (2 / P) * varia
```

```python
#define $b_k$
def b_k(n,P,y=10000):
    a, b = 0, P
    dx = (b-a)/y
    varia = 0
    for x in np.linspace(a,b,y):
        varia += f(x)*np.sin((n* np.pi*x)/P)
    varia *= dx
    return (2 / P) * varia
```

```python
def FourierSeries(x, P, n):
    a0 = a(0,P)
    sum = np.zeros(np.size(x))
    for i in np.arange(1, n + 1):
        sum += ((a_k(i, P)*np.cos((i*np.pi*x)/P)) + (b_k(i,P)*np.sin((i*np.pi*x)/P)))
    return (a0 / 2) + sum 
```

```python
#plots

plt.plot(x, f(x) , color = 'pink', label='signal')
#ST is Serir Trigonometric of Fourier
plt.plot(x, FourierSeries(x,P,6), color='Red', label='STF $1-x$, n=6')
plt.plot(x, FourierSeries(x,P,12), color='Blue', label='STF $1-x$, n=12')
plt.plot(x, FourierSeries(x,P,36), color='Green', label='STF $1-x$, n=36')
plt.plot(x, FourierSeries(x,P,124), color='Purple', label='STF $1-x$, n=124')

plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 2.2])

plt.legend()
plt.show()
```

$$ =2ax_1(-4x_1+8x_1x^2_2)+2bx_2(-12x^2_1x_2-6x_2)$$

```python
#model of the: differential equation ordinadiria DEO
def f(x1,x2):
    return -(4*x1) + (8*x1)*(x2**2)
def g(x1,x2):
    return -(12*x1**2)*(x2) - (6*x2)

def sistemaEQ(iv1, iv2, dt, time):
    Vx1.append(iv1)
    Vx2.append(iv2)
    for i in range(time):
        Vx1.append(Vx1[i] + (f(Vx1[i],Vx1[i])) * dt)
        Vx2.append(Vx2[i] + (g(Vx2[i],Vx2[i])) * dt)
    return Vx1, Vx2
```

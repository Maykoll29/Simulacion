# Monte Carlo Integration

#### Some integrands are unbounded and yet their integrals exist. Evaluate the integrals
$$\int\limits_{0}^{1} \frac{dx}{\sqrt{x}}$$
and
$$\int\limits_{0}^{1} \frac{sin(x)}{\sqrt{1-x}} dx$$
##### using the Monte Carlo methdo. Estimate the accuracy of your results for different sample sizes.

```python
#libraries
%matplotlib inline
import numpy as np
from sympy import *
from scipy import random
import matplotlib.pyplot as plt 
import math as mt
from math import sin
init_printing(use_unicode=True)
```

### a) 
$$\int\limits_{0}^{1} \frac{dx}{\sqrt{x}}$$

##### Real Value or Approximately:

```python
x = symbols('x')
integrate(1/sqrt(x), (x,0,1))
```

##### Using method Monte Carlo, for N=1000000

```python
a=0
b=1
N=1000000
rand=np.zeros(N)

for i in range(len(rand)):
    rand[i]=random.uniform(a,b)

def f(x):
    return 1/np.sqrt(x)

integral = 0.0

for i in range(N):
    integral += f(rand[i])
    
respuesta=(b-a)/float(N)*integral
    
respuesta
```

##### Plot of the function

```python
#init_printing(use_unicode=True)
%pylab inline
x = np.linspace(0, 1, 1000)
y = 1/ np.sqrt(x)
plt.figure(figsize=(6,5))
plt.plot(x, y)
plt.title("Funtion Real")
plt.show()
```

$N=100$

```python
N = 100


rec_x = [0,10]
rec_y = [-1,10]
# Sample from uniform distribution
mc_x = np.random.uniform(min(rec_x), max(rec_x), N)
mc_y = np.random.uniform(min(rec_y), max(rec_y), N)

# Define the points under the curve
points_under = [True if mc_y[i] <= f(mc_x[i]) else False 
                for i in range(len(mc_x))]

# Plot the points
plt.figure()
plt.plot(x, f(x), linewidth=2, c='k')
plt.scatter(mc_x[points_under], mc_y[points_under],
           c='green', s=1)
plt.scatter(mc_x[np.logical_not(points_under)], 
            mc_y[np.logical_not(points_under)], 
            c='pink', s=1)
plt.xlim(rec_x)
plt.ylim(rec_y)
plt.show()

# Calculate integral
integral = ((max(rec_x) - min(rec_x)) * 
            (max(rec_y) - min(rec_y)) * sum(points_under) / N)
```

## Pseudorandom number generators with linear structures

##### $a)$ Crack, i.e., find (m, a, c, x0), the following sequence 
$$-[ 61., 262., 151., 304., 785., 122., 427., 228., 613., 110., 831., 280., 569., 226., 339., 460., 653., 470., 999., 768.]$$
###### Provide the code used

Handmade Mathematical Form
$\alpha_1=(61,262)$

$\alpha_2 =(262,151)$

$\alpha_3=(151,304)$

$\alpha_4=(304,785)$

$\alpha_5=(...,...)$

...

$\alpha_n(n-1,n)$

$$
det\begin{pmatrix}
61 & 262 &1\\
262 & 151 &1\\
151 & 304 & 1\\
\end{pmatrix}=18432
$$
then we calculate the determinant of all the numbers of that vector formed by the alphas.
$$
det\begin{pmatrix}
653 & 470 &1\\
470 & 999 &1\\
999 & 768 & 1\\
\end{pmatrix}=68608
$$
So now your $gcd(18432,-76800,-332800,-292864,38912,77824,-48128,24576,-242688,-5232,105472,-84992,-54272,7168,59392,68608)=1024$,

then, $m=1024$ 

$\{61a+k=262,262a+k=151\}$ $mod (1024)$

$k=262-61a$ $mod (1024)$

$k=151-262a$ $mod (1024)$

$201a=-111$ $mod (1024)$

$a=137$ $mod(1024)$

$61(137)+k=262$ $mod(1024)$

$k=97$ $mod(1024)$

$a=137$

$c=k=97$

$m=1024$

```python
lista = [61, 262, 151, 304, 785, 122, 427, 228, 613, 110, 831, 280, 569, 226, 339, 460, 653, 470, 999, 768]
```

```python
def computer_terms(n,seed, a, c, m):
    numeros = [seed]
    for i in range(1,n):
        numeros.append((numeros[i-1]*a + c) % m)
    return numeros 
```

```python
a = 0
c = 0
m = 0
seed = 61
fun = False
for i in range(1,1000): 
    for j in range(1,1000):
        for k in range(999,1500): 
            num = computer_terms(20, seed, i, j, k)
            if num == lista:
                fun = True
                a = i
                c = j
                m = k
                break
        if fun == True:
            break
    if fun == True:
        break
if fun == True:
    print("a={}, c={}, m={}".format(a, c, m))
    encontrados = computer_terms(20,seed,a,c,m)
    print(encontrados)
else:
    print()
```
